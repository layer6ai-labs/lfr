```python
import numpy as np
import pandas as pd
```

```python
train_path = "../data/hepmass/1000_train.csv"
names = ['label']+[i for i in range(27)]
train_data = pd.read_csv(train_path, header=0, names=names)
train_data.head()
```

```python
test_path = "../data/hepmass/1000_test.csv"
names = ['label']+[i for i in range(27)]
test_data = pd.read_csv(test_path, header=0, names=names)
test_data.head()
```

```python
test_data = test_data.to_numpy()
train_data = train_data.to_numpy()

X_test = test_data[:, 1:]
y_test = test_data[:, 0].astype(np.int64)
X_train = train_data[:, 1:]
y_train = train_data[:, 0].astype(np.int64)
```

```python
X_train.shape, X_test.shape
```

```python
data_path = '../data/hepmass'
np.save( f'{data_path}/X_train', X_train)
np.save( f'{data_path}/X_test', X_test)
np.save( f'{data_path}/y_train', y_train)
np.save( f'{data_path}/y_test', y_test)
```

```python
from sklearn.metrics import accuracy_score as acc
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=200, metric='cosine')
knn.fit(X_train.reshape(X_train.shape[0],-1), y_train)
accuracy = knn.score(X_test.reshape(X_test.shape[0],-1), y_test)
accuracy
```

```python
import numpy as np
data_path = '../data/hepmass'
X_train=np.load( f'{data_path}/X_train.npy')
X_test=np.load( f'{data_path}/X_test.npy')
y_train=np.load( f'{data_path}/y_train.npy')
y_test=np.load( f'{data_path}/y_test.npy')
```

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
acc = []
for rand_seed in [0,1,2,3,4]:
    np.random.seed(rand_seed)
    lr = LogisticRegression(max_iter=300, random_state=rand_seed)
    lr.fit(X_train,y_train)
    predicted_labels = lr.predict(X_test)
    acc.append(metrics.accuracy_score(y_test, predicted_labels))
acc = np.stack(acc)
print(acc.mean(), acc.std())
```
